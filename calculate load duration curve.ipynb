Calculate load duration curve 


```python
import pandas as pd
import numpy as np
import json
import os
import matplotlib.pyplot as plt

# plr.rcParams['figure.figsize']=8,7
```

```python
os.chdir("raw_files")
```

```python
area_name='N1'
sub_name='NIB'

with open('dict_{}_{}.json'.format(area_name,sub_name), 'r') as f:
    dict_sub = json.load(f)
list_feeder=list(dict_sub.keys())
for remove_item in ['TP1', 'TP2', 'INC01', 'INC02', 'CAP1', 'CAP2', 'LINE01', 'LINE02']:
    list_feeder.remove(remove_item)

res_dict={'{}'.format(sub_name):{feeder_name:{'P_MW':{date:[dict_sub[feeder_name]['P_MW'][date][item] for item in ['{}'.format(item) for item in range(0,len(dict_sub[feeder_name]['P_MW'][date]))] if len(dict_sub[feeder_name]['P_MW'][date])==48] for date in list(dict_sub[feeder_name]['P_MW'].keys())}} for feeder_name in list_feeder}}
```

```python
num_feeder=0
x=list(range(0,48))
fig =plt.figure(figsize=(20, 40))
for feeder in list(res_dict['{}'.format(sub_name)].keys()):
    num_feeder=num_feeder+1
    ax=fig.add_subplot(5, 2, num_feeder)
    for date in list(res_dict['{}'.format(sub_name)][feeder]['P_MW'].keys()):
        if len(res_dict['{}'.format(sub_name)][feeder]['P_MW'][date])==48:
            ax.plot(x, res_dict['{}'.format(sub_name)][feeder]['P_MW'][date])
            ax.set_xlabel('t (30 minutes basis)')
            ax.set_ylabel('MW')
            ax.set_title('feeder {}'.format(feeder))
            ax.set_xticks(range(0,48))
plt.show()
```

```python
num_feeder=0
x=list(range(0,48))
fig =plt.figure(figsize=(20, 40))

for feeder in list(res_dict['{}'.format(sub_name)].keys()):
    num_feeder=num_feeder+1
#     print(feeder)
    ax=fig.add_subplot(5, 2, num_feeder)
    
    
    for date in list(res_dict['{}'.format(sub_name)][feeder]['P_MW'].keys()):
#         print(date)
        res_dict['{}'.format(sub_name)][feeder]['P_MW'][date].sort(reverse=True)

        if len(res_dict['{}'.format(sub_name)][feeder]['P_MW'][date])==48:
            ax.plot(x, res_dict['{}'.format(sub_name)][feeder]['P_MW'][date])
            ax.set_xlabel('t (30 minutes basis)')
            ax.set_ylabel('MW')
            ax.set_title('feeder {}'.format(feeder))
            ax.set_xticks(range(0,48))
plt.show()
```
