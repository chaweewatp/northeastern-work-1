```python
def get_half_hour_value(list_quarter_mins):
    sum_kWHr=0
    list_half_hour=[]
    for ind, i in enumerate(list_quarter_mins):
        if ind % 2 ==1:
            sum_kWHr=sum_kWHr+i
            list_half_hour.append(sum_kWHr)
            sum_kWHr=0
        else:
            sum_kWHr=sum_kWHr+i


    return list_half_hour
```

```python
get_half_hour_value([-10,20,30,40,100,120])
```
