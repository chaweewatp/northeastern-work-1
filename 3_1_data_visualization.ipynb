```python
x = [x for x in range(0,48,1)]
for feeder in list(dict_1.keys()):
    for day_type in ['working_day','non_working_day']:
        fig, ax = plt.subplots()
        for q1,q2,c in zip(['Quantile_0.0', 'Quantile_0.05', 'Quantile_0.1',  'Quantile_0.2',
                            'Quantile_0.3', 'Quantile_0.4'], 
                         ['Quantile_1.0', 'Quantile_0.95', 'Quantile_0.8',  'Quantile_0.7',
                          'Quantile_0.6', 'Quantile_0.5'],
                        ['yellow', 'red', 'blue', 'green','orange', 'purple']):
            y1=dict_1[feeder][day_type][q1]
            y2=dict_1[feeder][day_type][q2]
            ax.plot(x, y1, x, y2, color='black')
            ax.fill_between(x, y1, y2, facecolor=c)
            ax.set_title('{}, {}'.format(feeder,day_type))
        plt.savefig('{}_{}.png'.format(feeder,day_type))
        plt.show()
```

```python
#define load factor is average {(quantile_0.5 at t i )/(quantile_0.9 at t i )}
dict_AMR_LF={}
for key1 in list(dict_2.keys()):
    avg_load=dict_2[key1]['working_day']['Quantile_0.5'].mean()
    max_load=dict_2[key1]['working_day']['Quantile_0.95'].max()
    if max_load != 0:
        print('customer id {} has load factor = {}'.format(key1,avg_load/max_load))
        dict_AMR_LF.update({key1:{'max_load':max_load, 'LF':avg_load/max_load}})    
```

```python
dict_AMR_LF
```

```python
# plt.style.use('dark_background')
plt.style.use('ggplot')


fig, ax = plt.subplots()
LF=[dict_AMR_LF[x]['LF'] for x in dict_AMR_LF]
max_load=[dict_AMR_LF[x]['max_load'] for x in dict_AMR_LF]
ax.scatter(LF,max_load, c=LF, s=200)

# add annotate
name=[x for x in dict_AMR_LF]
for i, txt in enumerate(name):
    ax.annotate(txt, (LF[i], max_load[i]))

#add axis labe
ax.set_xlabel('Load factor')
ax.set_ylabel('max load MW')
ax.set_title('max load vs load factor TOU customers')
plt.savefig('images/scatter_max_load_vs_LF')
plt.show()
```
