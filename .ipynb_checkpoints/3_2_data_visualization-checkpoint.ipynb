```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from pylab import rcParams
plt.rcParams['figure.figsize'] = 20, 10
import matplotlib.gridspec as gridspec
plt.rcParams.update({'font.size': 22})
```

```python
#load npy files
dict_1=np.load('NIA_CI_dict.npy')[()]
```

```python
dict_1['SUB_OUT07']
```

```python
#see effect of CI
for feeder in list(dict_1.keys()):
    print(feeder)
    fig, ax = plt.subplots()
    
```

```python
x = [x for x in range(0,48,1)]
for feeder in list(dict_1.keys()):
    fig, ax = plt.subplots()
    for q1,q2,c in zip(['Quantile_0.0', 'Quantile_0.05', 'Quantile_0.1',  'Quantile_0.2'], 
                     ['Quantile_1.0', 'Quantile_0.95', 'Quantile_0.8',  'Quantile_0.7'],
                    ['yellow', 'red', 'blue', 'green']):
        y1=dict_1[feeder]['working_day'][q1]
        y2=dict_1[feeder]['working_day'][q2]
        ax.plot(x, y1, x, y2, color='black')
        ax.fill_between(x, y1, y2, facecolor=c)
        ax.set_title(feeder)
    plt.show()
```
