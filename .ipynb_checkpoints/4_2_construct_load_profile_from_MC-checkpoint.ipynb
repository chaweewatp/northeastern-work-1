```python
gridspec.GridSpec(1,5)
def plot_kWhr(dict_AMR, AMR_id):
    num_sample=20
    LF_data=[]
    fig = plt.figure(1)
    ax1 = plt.subplot2grid((1,5), (0,0), colspan=5, rowspan=1)

    for sample in np.arange(0,num_sample,1):
        x=list(dict_AMR_MC[AMR_id].keys())
        y=[dict_AMR_MC[AMR_id][x][sample] for x in dict_AMR_MC[AMR_id]]
        LF_data.append(find_LF(y))
        ax1.plot(x,y)
        
    ax2 = plt.subplot2grid((1,5), (0,1), colspan=1, rowspan=1)
    ax2.boxplot(LF_data)
    ax2.set_xticklabels(['{}'.format(AMR_id)])
    ax2.set_xlabel('AMR id')
    ax2.set_ylabel('Load factor')
    plt.show()

#     return LF_data
```
